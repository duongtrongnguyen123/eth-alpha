# 01 — EDA

Exploratory data analysis of raw OHLCV data.

```python
import pandas as pd
import plotly.graph_objects as go
from pathlib import Path

DATA_PATH = Path('..') / 'data' / 'ETHUSDT.csv'
df_raw = pd.read_csv(DATA_PATH, parse_dates=['timestamp'], index_col='timestamp')
df_raw.head()
```

```python
df_raw.info()
```

```python
df_raw.describe()
```

```python
print(f"From: {df_raw.index.min()}")
print(f"To:   {df_raw.index.max()}")
print(f"Rows: {len(df_raw):,}")
print(f"Nulls:\n{df_raw.isnull().sum()}")
```

```python
# 1-bar return distribution
r1 = df_raw['close'].pct_change().dropna()
fig = go.Figure(go.Histogram(x=r1, nbinsx=100, name='return_1'))
fig.update_layout(title='30-min Return Distribution', template='plotly_dark')
fig.show()
```

```python
# Close price over time
fig = go.Figure(go.Scatter(x=df_raw.index, y=df_raw['close'], name='close'))
fig.update_layout(title='ETHUSDT Close Price', xaxis_title='Date',
                  yaxis_title='Price (USD)', template='plotly_dark')
fig.show()
```

```python
# Volume over time
fig = go.Figure(go.Scatter(x=df_raw.index, y=df_raw['volume'], name='volume'))
fig.update_layout(title='ETHUSDT Volume', xaxis_title='Date',
                  yaxis_title='Volume', template='plotly_dark')
fig.show()
```
